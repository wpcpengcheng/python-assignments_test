### Exercise 2
Ask the user for a temperature in Celsius using `input()`. Convert it to Fahrenheit.

Formula: `F = C * 9/5 + 32`

Print the result like: `25.0°C = 77.0°F`

```python
# Exercise 2: Celsius to Fahrenheit converter
celsius_str = input("Enter temperature in Celsius: ")

try:
    celsius = float(celsius_str)
    fahrenheit = celsius * 9/5 + 32
    print(f"{celsius}°C = {fahrenheit}°F")
except ValueError:
    print("Please enter a valid number for Celsius.")
```

### Exercise 3
Create three variables: `price = 19.99`, `quantity = 3`, `tax_rate = 0.08`

Calculate the total cost with tax and print it rounded to 2 decimal places.

Formula: `total = price * quantity * (1 + tax_rate)`

```python
# Exercise 3: Calculate total with tax
price = 19.99
quantity = 3
tax_rate = 0.08

total = price * quantity * (1 + tax_rate)
print(f"Total (with tax): ${total:.2f}")
```

### Exercise 5
Ask the user for their first name and last name using two `input()` calls.

Print:
- Full name
- Total number of characters (excluding spaces)
- Initials (first letter of each name, uppercase)

```python
# Exercise 5: Name processing
first_name = input("First name: ")
last_name = input("Last name: ")

full_name = f"{first_name} {last_name}"
char_count = len((first_name + last_name).replace(" ", ""))
initials = f"{first_name[:1].upper()}{last_name[:1].upper()}"

print("Full name:", full_name)
print("Number of characters (no spaces):", char_count)
print("Initials:", initials)
```

### Exercise 6
Given the string `info = "John,25,3.8,True"`, split it and extract:
- name (string)
- age (convert to int)
- gpa (convert to float)
- is_student (convert to bool — hint: `"True" == "True"` gives True)

Print each with its type.

```python
# Exercise 6: Parse a data string
info = "John,25,3.8,True"

parts = info.split(",")
name = parts[0]
age = int(parts[1])
gpa = float(parts[2])
is_student = parts[3] == "True"

print(name, type(name))
print(age, type(age))
print(gpa, type(gpa))
print(is_student, type(is_student))
```

### Exercise 8
Write a tip calculator:
- `bill = 85.50`
- `tip_percent = 20`
- `num_people = 4`

Calculate and print:
- Tip amount
- Total bill (with tip)
- Amount each person pays

Round all values to 2 decimal places.

```python
# Exercise 8: Tip calculator
bill = 85.50
tip_percent = 20
num_people = 4

tip_amount = bill * (tip_percent / 100)
total_bill = bill + tip_amount
per_person = total_bill / num_people

print(f"Tip amount: ${tip_amount:.2f}")
print(f"Total bill (with tip): ${total_bill:.2f}")
print(f"Amount per person: ${per_person:.2f}")
```
